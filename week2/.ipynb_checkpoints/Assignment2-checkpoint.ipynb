* **value** : data value for element (tenths of degrees C)

```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
import matplotlib.pyplot as plt
#import mplleaflet
import pandas as pd
from datetime import date
import numpy as np

df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
df.head()
```

```python
years=pd.DatetimeIndex(df['Date']).year
months=pd.DatetimeIndex(df['Date']).month
days=pd.DatetimeIndex(df['Date']).day
df['Year']=years
df['Month']=months
df['Day']=days
df['Data_Value']=df['Data_Value']/10
df.drop(['ID','Date'],1, inplace=True)
df.head()
```

```python
df1=df[(df['Month']!=2) & (df['Day']!=29)]
df2=df1[df1['Year']<2015]
df_2015=df[df['Year']>=2015]
df2
#df.drop(leap_year.index)
```

```python
df_min=df2[df2['Element']=='TMIN'].groupby(['Month','Day']).aggregate({'Data_Value':np.min}).reset_index()
df_max=df2[df2['Element']=='TMAX'].groupby(['Month','Day']).aggregate({'Data_Value':np.max}).reset_index()
df_min_2015=df_2015[df_2015['Element']=='TMIN'].groupby(['Month','Day']).aggregate({'Data_Value':np.min}).reset_index()
df_max_2015=df_2015[df_2015['Element']=='TMAX'].groupby(['Month','Day']).aggregate({'Data_Value':np.max}).reset_index()
df_max
```

```python
#df_brokenRecord_min =df_min_2015[df_min_2015['Data_Value'] > (df_min_2015['Month']).map(df_min['Data_Value'])]
df3_min=pd.merge(df_min, df_min_2015, how='inner', on=['Month','Day'])
df3_max=pd.merge(df_max, df_max_2015, how='inner', on=['Month','Day'])
df_brokenRecord_min=df3_min[df3_min['Data_Value_x']>df3_min['Data_Value_y']]
df_brokenRecord_max=df3_max[df3_max['Data_Value_x']<df3_max['Data_Value_y']]
df3_min
```

```python
mins_values=df_min['Data_Value'].tolist()
mins_months=df_min['Month'].tolist()
mins_days=df_min['Day'].tolist()
mins_axis=[]

maxs_values=df_max['Data_Value'].tolist()
maxs_months=df_max['Month'].tolist()
maxs_days=df_max['Day'].tolist()
maxs_axis=[]

for i in range(len(mins_values)):
    mins_axis.append((date(2015,mins_months[i],mins_days[i] ) - date(2015, 1, 1)).days)
    
for i in range(len(maxs_values)):
    maxs_axis.append((date(2015,maxs_months[i],maxs_days[i] ) - date(2015, 1, 1)).days)  
```

```python
df_brokenRecord_min.drop(['Data_Value_x'],1, inplace=True)
df_brokenRecord_max.drop(['Data_Value_x'],1, inplace=True)
df_brokenRecord_min.rename(columns={'Data_Value_y': 'Data_Value'}, inplace=True)
df_brokenRecord_max.rename(columns={'Data_Value_y': 'Data_Value'}, inplace=True)
```

```python
from datetime import date
mins_brokenRecord_values=df_brokenRecord_min['Data_Value'].tolist()
mins_brokenRecord_months=df_brokenRecord_min['Month'].tolist()
mins_brokenRecord_days=df_brokenRecord_min['Day'].tolist()
mins_brokenRecord_axis=[]

maxs_brokenRecord_values=df_brokenRecord_max['Data_Value'].tolist()
maxs_brokenRecord_months=df_brokenRecord_max['Month'].tolist()
maxs_brokenRecord_days=df_brokenRecord_max['Day'].tolist()
maxs_brokenRecord_axis=[]

for i in range(len(mins_brokenRecord_values)):
    mins_brokenRecord_axis.append((date(2015,mins_brokenRecord_months[i],mins_brokenRecord_days[i] ) - date(2015, 1, 1)).days)
    
for i in range(len(maxs_brokenRecord_values)):
    maxs_brokenRecord_axis.append((date(2015,maxs_brokenRecord_months[i],maxs_brokenRecord_days[i] ) - date(2015, 1, 1)).days)    
```

```python
plt.figure(figsize=(10,8))
colors = ['green', 'red']

plt.plot(mins_axis,mins_values, c='green', alpha = 0.3, label = 'Minimum Temperature (2005-2014)')
plt.plot(maxs_axis,maxs_values, c ='red', alpha = 0.3, label = 'Maximum Temperature (2005-2014)')
plt.scatter(mins_brokenRecord_axis, mins_brokenRecord_values, s = 8, c = 'blue', label = 'Record Break Minimum (2015)')
plt.scatter(maxs_brokenRecord_axis, maxs_brokenRecord_values, s = 8, c = 'darkred', label = 'Record Break Maximum (2015)')
plt.fill_between(mins_axis, mins_values, maxs_values, facecolor='grey', alpha=0.2)
plt.ylim(-45, 60)
plt.legend(loc ='best', frameon=False,fontsize=10)
plt.xticks( np.linspace(0, 30*11 , num = 12), (r'Jan', r'Feb', r'Mar', r'Apr', r'May', r'Jun', r'Jul', r'Aug', r'Sep', r'Oct', r'Nov', r'Dec') )
plt.xlabel('Months',fontsize=12)
plt.ylabel('Temperature (tenths of degrees C')
plt.title('2015 temperature broke records vs (2005-2014) temperature records (°C)')
plt.show()
```
